#### Energy Minimization 
Starting from the intial configuration, ensure that the system has no steric clashes or non real geometry. Minimizes the free potential energy and force to less than Fmax

```python
import matplotlib.pyplot as plt
import panedr
```

```python
path = 'data/em.edr'
df = panedr.edr_to_df(path=path)
list(df.columns)
```

```python
time = df['Time']
potential_energy = df['Potential']
plt.plot(time, potential_energy)
plt.title("Energy Minimzation \n 1AKI, Steepest Descent")
plt.xlabel("Time (ps)")
plt.ylabel("Potential Energy (kj/mol)")
plt.savefig("plots/Energy_minmization.png")
plt.show()
```

```python
path = 'data/nvt.edr'
df = panedr.edr_to_df(path=path)
list(df.columns)
```

```python
time = df['Time']
potential_energy = df['Temperature']
plt.plot(time, potential_energy)
plt.title("Temperture Over Time \n 1AKI, NVT Ensemble")
plt.xlabel("Time (ps)")
plt.ylabel("Temperture (K)")
plt.savefig("plots/Temperture_Stab.png")
plt.show()
```

```python
path = 'data/npt.edr'
df = panedr.edr_to_df(path=path)
list(df.columns)
```

```python
time = df['Time']
pressure = df['Pressure']
pressure_average = pressure.rolling(window=10).mean()

plt.plot(time, pressure)
plt.plot(time, pressure_average, label='10 ps rolling avg')
plt.title("Pressure Over Time \n 1AKI, NPT Ensemble")
plt.xlabel("Time (ps)")
plt.ylabel("Pressure (bar)")
plt.legend()
plt.savefig("plots/Pressure_Stab.png")
plt.show()
```

```python
density = df['Density']
density_average = density.rolling(window=10).mean()

plt.plot(time, density)
plt.plot(time, density_average, label='10 ps rolling avg')
plt.title("Density Over Time \n 1AKI, NPT Ensemble")
plt.xlabel("Time (ps)")
plt.ylabel("Density (kg/m^3)")
plt.legend()
plt.savefig("plots/Density_Stab.png")
plt.show()
```
